```python
#subtask 1 and 2
def calculate_evaporation(temp=37,humid=45):
 '''a function to calculate evaporation 

 parameters:

 1. temp in celcius
 
 2. humid in percentage'''
 return 0.5*(temp+25)*(100-humid)
```

```python
print(calculate_evaporation())
print(calculate_evaporation(20,30))
print(calculate_evaporation(10)) #took temp as 10
print(calculate_evaporation(temp=calculate_evaporation.__defaults__[0],humid=50))#kept temp as the default value and changed humidity.
```
